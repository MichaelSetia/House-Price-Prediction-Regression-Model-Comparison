# Missing Value ( Data Kosong )
----------------------------------------------------------

```python
import pandas as pd

# 1. Load Data (sesuaikan jika datanya bentuk lain pd.read_nama file)
train = pd.read_csv("train.csv") 

# 2. Ngitung Jumlah Missing Value Setiap Colom
missing_values = train.isnull().sum()

# 3. Pisahkan kolom yang null nya di atas 1000 dan di bawah 1000
less = missing_values[missing_values < 1000].index
over = missing_values[missing_values >= 1000].index

# 4. Ambil kolom numerik dan isi data yang kosong pada semua kolom numerik dengan median dari kolom tersebut
numeric_feature = train[less].select_dtypes(include=["number"]).columns
train[numeric_feature] = train[numeric_feature].fillna(train[numeric_feature].median())

# 5. Ambil kolom string dan isi dengan modus value
object_feature = train[less].select_dtypes(include=["object","string"]).columns
for colum in object_feature:
    train[colum] = train[colum].fillna(train[colum].mode()[0])

# 6. Buang kolom yang bolongnya >= 1000
df = train.drop(columns=over)

# 7. Cek hasil akhir 
missing_values_clean = df.isnull().sum()
print("Kolom yang masih memiliki missing values:")
print(missing_values_clean[missing_values_clean > 0])
```

# Outlier Data (Data yang nilainya jauh dari data lainnya)
----------------------------------------------------------

```python
# Menghitung IQR, Q1, dan Q3
# Q1 (Quartile 1): Nilai di persentil ke-25 data.
# Q3 (Quartile 3): Nilai di persentil ke-75 data.
# IQR: Rentang antara Q3 dan Q1 (IQR = Q3 - Q1).


Q1 = df[numeric_feature].quantile(0.25)
Q3 = df[numeric_feature].quantile(0.75)
IQR = Q3 - Q1

print(IQR)

# Menentukan Batas Bawah dan Batas Atas
# Batas Bawah: Q1 - 1.5 * IQR
# Batas Atas: Q3 + 1.5 * IQR

condition = ~((df[numeric_feature] < (Q1 -1.5*IQR)) | (df[numeric_feature] > (Q3+1.5*IQR))).any(axis=1) 
df_filter_numeric = df.loc[condition,numeric_feature]


categorical_feature = df.select_dtypes(include=['object','string']).columns
df= pd.concat([df_filter_numeric,df.loc[condition,categorical_feature]],axis=1)

df.head()
```

```python
import sklearn

x = df_lencoder.drop(columns=["SalePrice"])
y = df_lencoder["SalePrice"]

print(y)


```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
print("Jumlah data: ",len(x))
# menghitung panjang/jumlah data pada x_train
print("Jumlah data latih: ",len(x_train))
# menghitung panjang/jumlah data pada x_test
print("Jumlah data test: ",len(x_test))
```

```python
from sklearn import linear_model
import sklearn.linear_model

lars = linear_model.Lars(n_nonzero_coefs=1).fit(x_train,y_train)


from sklearn.linear_model import LinearRegression
LR= LinearRegression().fit(x_train,y_train)


from sklearn.ensemble import GradientBoostingRegressor
GBR = GradientBoostingRegressor(random_state=184)
GBR.fit(x_train,y_train)

```

```python
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error

pred_lars = lars.predict(x_test)
mae_lars = mean_absolute_error(y_test,pred_lars) 
mse_lars = mean_squared_error( y_test, pred_lars)
r2_lars =r2_score(y_test,pred_lars)

data = {
    "MAE" : mae_lars,
    "MSE" : mse_lars,
    "R2" : r2_lars
}

df_result = pd.DataFrame(data,index=["Lars"])
df_result


```

```python
pred_LR = LR.predict(x_test)
mae_LR = mean_absolute_error(y_test,pred_LR) 
mse_LR = mean_squared_error( y_test, pred_LR)
r2_LR =r2_score(y_test,pred_LR)

df_result.loc["Linear Regression"] = [mae_LR,mse_LR,r2_LR]
df_result
```

```python
pred_GBR = GBR.predict(x_test)
mae_GBR = mean_absolute_error(y_test,pred_GBR)
mse_GBR = mean_squared_error(y_test,pred_GBR)
r2_GBR = r2_score(y_test,pred_GBR)

df_result.loc["Gradient Boosting Regressor"] = [mae_GBR,mse_GBR,r2_GBR]
df_result
```
